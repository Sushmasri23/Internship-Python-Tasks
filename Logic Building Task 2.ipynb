# Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

words = sentence.split()        # Split sentence into words
unique_words = set(words)       # Remove duplicates

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

max_employee = None
max_salary = 0

for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        max_employee = name

print("Highest Salary:", max_employee, "-", max_salary)
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

# Problem 4: Count Products Above Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# Problem 5: Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

# Problem 7: Count Character Frequency

```python
text = "python"

freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

# Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

total = sum(marks)
count = len(marks)

average = total / count

print("Average Marks:", average)
```
