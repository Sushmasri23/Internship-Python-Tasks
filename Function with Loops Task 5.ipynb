# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes = sum(likes)
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

```python
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)

check_stock(6)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_check(data, required=70):
    avg = sum(data) / len(data)
    
    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_check([60, 80, 70, 75])
```

# Problem 4: Duplicate Account Detection

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user2", "user1"]
check_duplicates(users)
```

# Problem 6: Premium Crop Price Filter

```python
def premium_crops(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)

premium_crops([1500, 2500, 3200, 1800])
```

# Problem 7: System Monitoring – Application Health Checker

```python
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

system_health(7)
```

# Problem 9: Student Attendance Eligibility

```python
def attendance_check(attendance):
    present = attendance.count("P")
    percentage = (present / len(attendance)) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_check(["P", "P", "A", "P", "P"])
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Category:", status)

electricity_bill(250)
```
