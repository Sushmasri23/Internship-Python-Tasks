# Problem 1: Smart Parking Lot Management System

```python
def parking_system(capacity, logs):
    parked = 0
    max_parked = 0
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        if parked > max_parked:
            max_parked = parked
    
    print("Currently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

# Problem 2: Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base time (5 min per km)
    
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    
    print("Estimated Delivery Time:", time, "minutes")

delivery_time(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def occupancy(total_seats, booked_list):
    booked = len(booked_list)
    
    percentage = (booked / total_seats) * 100
    
    print("Occupancy:", int(percentage), "%")
    
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

occupancy(200, [1]*150)
```

# Problem 4: Cloud Server Load Classification

```python
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    
    print("Average CPU Load:", int(avg), "%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

server_load([45, 60, 70, 85, 90])
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):
    overused = []
    
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    
    print("Overused Resources:", ", ".join(overused) if overused else "None")
    
    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

# Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")

event_registration(100, 105)
```
